### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 11)
y = x**2
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x²')
plt.title('Line Plot of y = x²')
plt.xlabel('x values')
plt.ylabel('y values (x squared)')
plt.grid(True)
plt.legend()
plt.show()
```
