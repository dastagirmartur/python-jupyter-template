### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1, 6)
y = x ** 2
categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]
scatter_y = [2, 4, 6, 8, 10]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(x, y, marker='o')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y = x²')
axs[0, 0].grid(True)
axs[0, 1].bar(categories, bar_values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Category')
axs[0, 1].set_ylabel('Value')
axs[1, 0].scatter(x, scatter_y, color='purple')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')
plt.tight_layout()
plt.show()
```
